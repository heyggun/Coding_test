```python
from collections import deque

def solution(m, n, board):
    answer= 0
    board = [list(_) for _ in board]
    tmp_set = set()

    while True:
        for i in range(m-1):
            for j in range(n-1):
                b = board[i][j]
                if b ==[]:
                    continue
                if board[i][j+1] == b and board[i+1][j] ==b and board[i+1][j+1] == b:
                    tmp_set.add((i,j))
                    tmp_set.add((i,j+1))
                    tmp_set.add((i+1,j))
                    tmp_set.add((i+1,j+1))
        if tmp_set:
            answer += len(tmp_set)
            for t_i,t_j in tmp_set:
                board[t_i][t_j] = []
            tmp_set = set()
        else:
            return answer

        while True:
            moved = 0
            for i in range(m-1):
                for j in range(n):
                    if board[i][j] and board[i+1][j] == []:
                        board[i+1][j] = board[i][j]
                        board[i][j] = []
                        moved = 1
            if moved==0:
                break
              
```
