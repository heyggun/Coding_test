**입출력 예**



**입출력 예 설명**

    입출력 예 #1
    답을 도출하는 과정은 다음과 같습니다.



    따라서 7을 return 해야 합니다.
    입출력 예 #2

    답을 도출하는 과정은 다음과 같습니다.



    따라서 229를 return 해야 합니다.

```python
def solution(n):
    answer = 0
    remainder = ''
    while n//3 !=0:
        remainder += str(n%3)
        n = n//3
        
    remainder += str(n%3)
    remainder = remainder[::-1]
    
    for i in range(len(remainder)):
        answer += int(remainder[i]) * (3**i)
    
    return answer
```
