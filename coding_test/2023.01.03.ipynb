```python
#    :  O(N)


def getDivisor(n):
    n = int(n)
    divisor = list()
    
    for i in range(1, n+1):
        if n%i == 0:
            divisor.append(i)
            
    return divisor
```

```python
getDivisor(5)
```

```python
#    :  O(N^(1/2)))

def getDivisor2(n):
    divisor = list()   
    for i in range(1, int(n**(1/2))+1):
        if (n%i)==0:
            divisor.append(i)
            if (i**2 !=n):
                divisor.append(n//i)
    divisor.sort()
    
    return divisor
```

```python
getDivisor(24)
```

```python
getDivisor2(24)
```

```python
def solution(number, limit, power):
    divisor_lst = list()
    answer_lst = list()
    
    for num in range(1,number+1):
        cnt = 0
        for i in range(1,num+1):
            if num%i==0:
                cnt+=1
        divisor_lst.append(cnt)    
   
    for divisor in divisor_lst:
        if divisor > limit:
            answer_lst.append(power)
        else:
            answer_lst.append(divisor)
        
    answer = sum(answer_lst)
    return answer
```

```python
print(solution(5,3,2))
print(solution(10,3,2))
```

```python
def solution(k, m, score):
    pakage = 0
    answer = 0
    
    score.sort(reverse=True)
    
    for s in score:
        if s>k:
            score.remove(s)
            score.append(s)
    
    package = len(score)//m
    start = 0
    for i in range(1, package+1):
        tmp_score = score[start:m*i]
        answer += min(tmp_score) * m
        start +=m
   
    return answer
```

```python
def solution(food):
    answer = ''
    tmp_lst, food_lst, total_food_lst = list(), list(), list()
    
    for i in range(1,4):
        cnt = food[i]//2
        tmp_lst = [i] * cnt
        food_lst = food_lst + tmp_lst
        
    reverse_food_lst = food_lst.copy()
    reverse_food_lst.sort(reverse=True)
    
    total_food_lst = food_lst + [0] + reverse_food_lst
    answer = ''.join(map(str,total_food_lst))
    
    return answer
```

```python
print(solution([1,3,4,6]))
print(solution([1,7,1,2]))
```
