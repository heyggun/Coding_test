```python
def solution(X, Y):
    answer = ''
    
    for i in range(9,-1,-1):
        answer += (str(i) * min(X.count(str(i)), Y.count(str(i))))
    
    if answer == '':
        result = '-1'
    elif len(answer) == answer.count('0'):
        result = '0'
    else:
        result = answer
    
    return result
```

```python
print(solution("100", "2345"))
print(solution("100", "203045"))
print(solution("100", "123450"))
print(solution("12321", "42531"))
print(solution("5525", "1255"))
```

    따라서 "RCJA"를 return 해야 합니다.

```python
def solution(survey, choices):
    score_dict = {1:3, 2:2, 3:1, 4:0, 5: 1, 6:2, 7:3}
    char_dict = {'R':0, 'T':0,'C':0, 'F':0,'J':0, 'M':0, 'A':0, 'N':0}
    answer = ''
    
    for idx, choice in enumerate(choices):
        if choice <=3:
            char = survey[idx][0]
            char_dict[char] += score_dict[choice]

        elif choice >= 5:
            char = survey[idx][1]
            char_dict[char] += score_dict[choice]

        else:
            pass


    if char_dict['R'] >= char_dict['T']:
        answer += 'R'
    else: 
        answer += 'T'

    if char_dict['C'] >= char_dict['F']:
        answer+='C'
    else:
        answer+='F'

    if char_dict['J'] >= char_dict['M']:
        answer+='J'
    else:
        answer+='M'

    if char_dict['A'] >= char_dict['N']:
        answer +='A'
    else:
        answer+='N'
    return answer
```

```python
print(solution(["AN", "CF", "MJ", "RT", "NA"],[5, 3, 2, 7, 5]))
print(solution(["TR", "RT", "TR"],[7, 1, 3]))
```
