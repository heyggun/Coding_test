```python
def solution(s):
    answer = ''
    for string in s.split(' '):
        for idx in range(len(string)):
            if idx%2==0:
                answer += string[idx].upper()
            else:
                answer += string[idx].lower()
        answer+= ' '
    return answer[:-1]
```

```python
print(solution("try hello world"))
```
