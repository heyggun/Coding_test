**문제 설명**

    어떤 문장의 각 알파벳을 일정한 거리만큼 밀어서 다른 알파벳으로 바꾸는 암호화 방식을 시저 암호라고 합니다. 예를 들어 "AB"는 1만큼 밀면 "BC"가 되고, 3만큼 밀면 "DE"가 됩니다. "z"는 1만큼 밀면 "a"가 됩니다. 문자열 s와 거리 n을 입력받아 s를 n만큼 민 암호문을 만드는 함수, solution을 완성해 보세요.
    
**제한 조건**

    공백은 아무리 밀어도 공백입니다.
    s는 알파벳 소문자, 대문자, 공백으로만 이루어져 있습니다.
    s의 길이는 8000이하입니다.
    n은 1 이상, 25이하인 자연수입니다.
    
    
**입출력 예**



```python
from string import ascii_uppercase, ascii_lowercase

def solution(s, n):
    answer = []
    
    for string in s:
        if string.isupper():
            num = list(ascii_uppercase).index(string)
            answer.append(ascii_uppercase[(num+n)%26])

        elif string.islower():
            num = list(ascii_lowercase).index(string)
            answer.append(ascii_lowercase[(num+n)%26])
        
        else:
            answer.append(' ')
    
    return ''.join(answer)
```

```python
print(solution("AB",1))
print(solution("z",1))
print(solution("a B z",4))
```

**다른 사람 풀이**

    - 유니코드 정수로 변환해주는 ord 함수를 사용한 풀이
    
    

```python
s = "a B z"
n = 4

string_lst = list(s)
print(string_lst)

for i in range(len(string_lst)):
    if string_lst[i].isupper():
        string_lst[i] = chr((ord(string_lst[i])- ord('A')+ n)%26 + ord('A'))
        
    elif string_lst[i].islower():
        string_lst[i]= chr((ord(string_lst[i]) - ord('a')+ n)%26 + ord('a'))
        
answer = ''.join(string_lst)
answer
```

    - 앞에서 %로 나머지를 구하면 0부터 25가 나오게 되는데 아스키코드의 a는 시작 값은 97 
    A의 시작 값은 65이기 때문에 더해주는 것입니다 값을 더해줌으로써 chr로 형변환 했을 때 알파벳이 저장되기 때문
