## 파이썬을 파이썬답게

**문제 설명**

    문자열을 담은 이차원 리스트, mylist 가 solution 함수의 파라미터로 주어집니다. solution 함수가 mylist를 일차원 리스트로 만들어 리턴하도록 코드를 작성해주세요.
    
**제한 조건**

    arr의 길이는 1 이상 100 이하인 자연수입니다.
    arr 원소의 길이는 5를 넘지 않습니다.
    
**예시**



```python
def solution(mylist):
    answer = []
    for idx in range(len(mylist)):
        answer+= mylist[idx]
    return answer
```

```python
print(solution([[1],[2]]))
print(solution([['A','B'], ['X','Y'],['1']]))
```

```python
import numpy as np

mylist = [['A','B'], ['X','Y'],['1']]
answer = []
for idx in range(len(mylist)):
    answer+=mylist[idx]
    
answer
```

    -> 그냥 for문 idx 사용 안하고.. 리스트로 그냥 += 해도 됐음

```python
answer = []

for i in mylist:
    answer+=i
answer
```

`파이썬을 파이썬 답게`


```python
# sum 함수

answer = sum(mylist, [])
answer
```

```python
# itertools.chain

import itertools

list(itertools.chain.from_iterable(mylist))
```

```python
#itertools & unpacking

list(itertools.chain(*mylist))
```

```python
#list comprehension

[element for array in mylist for element in array]
```

```python
# reduce 함수

from functools import reduce

list(reduce(lambda x, y: x+y, mylist))
```

```python
# reduce 함수 이용

from functools import reduce
import operator

list(reduce(operator.add, mylist))
```
