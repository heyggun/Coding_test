**문제 설명**

    숫자를 담은 일차원 리스트, mylist에 대해 mylist의 원소로 이루어진 모든 순열을 사전순으로 리턴하는 함수 solution을 완성해주세요.

**제한 조건**

    mylist 의 길이는 1 이상 100 이하인 자연수입니다.
    
**예시**



```python
from itertools import permutations

def solution(mylist):
    mylist.sort()
    return list(map(list, [per for per in permutations(mylist, len(mylist))]))
```

```python
print([2,1])
print([1,2,3])
```

     - permutations 그대로 넣고 map 함수로 iterator 해도 됐다!

`python을 python답게`

    - 순열을 만들기 위해서 아래와 같은 작업~

```python
import itertools

lst = ['A','B','C']

list(map(''.join, itertools.permutations(lst, len(lst))))
```
