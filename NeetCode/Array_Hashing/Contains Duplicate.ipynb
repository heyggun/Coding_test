Array & Hashing

Given an integer array nums, return true if any value appears more than once in the array, otherwise return false.

Example 1:

Input: nums = [1, 2, 3, 3]

Output: true

Example 2:

Input: nums = [1, 2, 3, 4]

Output: false

```python
example_nums_1 = [1, 2, 3, 3] 
example_nums_2 = [1, 2, 3, 4] 
```

```python
from typing import List

class Solution:
    def hasDuplicate(self, nums: List[int]) -> bool:
        return len(nums) != len(set(nums))
        
```

```python
sol = Solution()
print(sol.hasDuplicate(example_nums_1))
print(sol.hasDuplicate(example_nums_2))
```

개선하기 - `early_stop`

```python
class Soultion:
    def hasDupliate(self, nums: List[int]) -> bool:
        seen = set()
        for n in nums:
            if n in seen:
                return True
            seen.add(n)
            
        return False
        
        
```

```python
sol = Solution()
print(sol.hasDuplicate(example_nums_1))
print(sol.hasDuplicate(example_nums_2))
```

공간 복잡도 줄여보기 - but 시간 복잡도는 올라감. Trade-off

```python
class Solution:
    def hasDuplicate(self, nums:List[int]) -> bool:
        nums.sort()
        for i in range(1, len(nums)):
            if nums[i] == nums[i-1]:
                return True
            
        return False
```

```python
sol = Solution()
print(sol.hasDuplicate(example_nums_1))
print(sol.hasDuplicate(example_nums_2))
```
