## Leetcode 1291. Sequential Digits

https://leetcode.com/problems/sequential-digits/description/?envType=daily-question&envId=2024-02-02

**Description**

    An integer has sequential digits if and only if each digit in the number is one more than the previous digit.

    Return a sorted list of all the integers in the range [low, high] inclusive that have sequential digits.

    

    Example 1:

    Input: low = 100, high = 300
    Output: [123,234]
    Example 2:

    Input: low = 1000, high = 13000
    Output: [1234,2345,3456,4567,5678,6789,12345]
 

**Constraints**

    10 <= low <= high <= 10^9


```python
from typing import List
```

```python
low = 100, high = 300
```

```python
class Solution:
    def sequentialDigits(self, low: int, high: int) -> List[int]:
        nums = '123456789'
        ans = []

        for i in range(len(str(low)), len(str(high))+1):
            for j in range(10-i):
                tmp = int(nums[j:j+i])
                if low<= tmp <= high:
                    ans.append(tmp)

        return ans
```

```python
sol = Solution()
print(sol.sequentialDigits(low = 100, high = 300))
print(sol.sequentialDigits(low = 1000, high = 13000))
```

```python
class Solution:
    def sequentialDigits(self, low: int, high: int) -> List[int]:
        ans = []

        for i in range(1,10):
            num = i
            nxt_num = i+1
            while num <= high and nxt_num<=9:
                num = num*10 + nxt_num
                if low<=num<=high:
                    ans.append(num)
                nxt_num+=1

        return sorted(ans)

            
```

```python
sol = Solution()
print(sol.sequentialDigits(low = 100, high = 300))
print(sol.sequentialDigits(low = 1000, high = 13000))
```
