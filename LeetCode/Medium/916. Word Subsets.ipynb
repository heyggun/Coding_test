### Leetcode 916. Word Subsets

https://leetcode.com/problems/word-subsets/description/?envType=daily-question&envId=2025-01-10

```python
from typing import List
```

```python
class Solution:
    def wordSubsets(self, words1: List[str], words2: List[str]) -> List[str]:
        ans = []
        max_freq = [0] * 26
        
        for word in words2:
            tmp_freq = [0] * 26
            for w in word:
                tmp_freq[ord(w)-ord('a')] +=1
                
            for i in range(26):
                max_freq[i] = max(max_freq[i], tmp_freq[i])
            
        for word in words1:
            freq = [0] * 26
            for w in word:
                freq[ord(w)-ord('a')] +=1
            
            if all(freq[i] >= max_freq[i] for i in range(26)):
                ans.append(word)
                
        return ans
```

```python
solution = Solution()
print(solution.wordSubsets(words1 = ["amazon","apple","facebook","google","leetcode"], words2 = ["e","o"]))
print(solution.wordSubsets(words1 = ["amazon","apple","facebook","google","leetcode"], words2 = ["l","e"]))
```
